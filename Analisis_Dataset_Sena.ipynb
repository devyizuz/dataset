```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/devyizuz/dataset/main/Georeferenciaci_n_Centros_de_Aprendizaje_SENA_20240730.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.tail()
```

```python
registros_con_comillas_dobles = df['CENTRO_FORMACION'].str.contains('"').sum()
registros_con_comillas_simples = df['CENTRO_FORMACION'].str.contains("'").sum()
total_registros_con_comillas = registros_con_comillas_dobles + registros_con_comillas_simples

total_registros_con_comillas
```

```python
registros_con_comas = df['CENTRO_FORMACION'].str.contains(',').sum()

registros_con_comas
```

```python
df['CENTRO_FORMACION'] = df['CENTRO_FORMACION'].str.replace(',', '').str.replace("'", '')
```

```python
registros_con_comas = df['CENTRO_FORMACION'].str.contains(',').sum()

registros_con_comas
```

```python
registros_con_punto_comas = df['CENTRO_FORMACION'].str.contains(';').sum()

registros_con_punto_comas
```

```python
duplicados = df[df.duplicated()]
print("Filas duplicadas:")
print(duplicados)
```

```python
df_sin_duplicados = df.drop_duplicates()
```

```python
print("DataFrame sin duplicados:")
print(df_sin_duplicados)
```

```python
num_registros = df_sin_duplicados.shape[0]
print(f"El número de registros en el dataset es: {num_registros}")
```

```python
numero_centros_por_departamento = df_sin_duplicados.groupby('NOMBRE_DEPARTAMENTO')['CENTRO_FORMACION'].count()
```

```python
numero_centros_ordenados = numero_centros_por_departamento.sort_values(ascending=False)
print(numero_centros_ordenados)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
numero_centros_ordenados.plot(kind='bar', color='skyblue')
plt.title('Número de Centros de Aprendizaje por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de Centros')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

Distribución por departamento:

    La gráfica de barras muestra la cantidad de centros de aprendizaje por departamento. Se observa una mayor concentración en departamentos como Antioquia y Bogotá D.C., lo que sugiere una tendencia a concentrar más centros en áreas más pobladas o de mayor actividad económica.
